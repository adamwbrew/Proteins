```python
import pandas as pd
```

```python
protiens_2 = pd.read_excel("Adam_Proteomics_2.xlsx", sheet_name=0)
protiens_identifiers_2 = pd.read_excel("Adam_Proteomics_2.xlsx", sheet_name=1)
```

```python
protiens_2.head()
```

```python
protiens_2.info()
```

```python
# remove unwanted columns
protiens_2 = protiens_2[['Protein.Group', 'Protein.Ids','Protein.Names','Genes','First.Protein.Description',1139,3058]]
protiens_2.head()
```

```python
protiens_2.info()
```

```python
# if NaN in columns 1139 or 3058 change to zero per request
protiens_2 = protiens_2.fillna(0)
protiens_2 = protiens_2.rename(columns= {"Protein.Names":"Protein_Names",
                                     "Protein.Group":"Protein_Group",
                                     "Protein.Ids":"Protein_Ids",
                                     "First.Protein.Description":"First_Protein_Description",
                                     1139:"1139",
                                     3058:"3058"})
protiens_2["Peptides"] = None
protiens_2.info()
```

```python
protiens_2.head()
```

```python
protiens_identifiers_2 = protiens_identifiers_2.rename(columns = {"Sample":"Sample_Name"})
protiens_identifiers_2.head()
```

```python
protiens_identifiers_2.info()
```

```python
protiens_2.to_csv("proteins_denormalized_2.csv")
protiens_identifiers_2.to_csv("proteins_identifiers_denormalized_2.csv")
```

```python
# create mouse to protiens hit table based on protiens_2
protein_mouse_1 = protiens_2[["Protein_Names", "1139"]]
protein_mouse_2 = protiens_2[["Protein_Names", "3058"]]

protein_mouse_1["Sample_Name"] = "1139"
protein_mouse_2["Sample_Name"] = "3058"

protein_mouse_1 = protein_mouse_1.rename(columns = {"1139":"Hits"})
protein_mouse_2 = protein_mouse_2.rename(columns = {"3058":"Hits"})

union_dfs = pd.concat([protein_mouse_1,
                       protein_mouse_2], ignore_index=True)

union_dfs = union_dfs[["Protein_Names", "Sample_Name", "Hits"]]
union_dfs.head()
```

```python
union_dfs.info()
```

```python
union_dfs.to_csv("mouse_by_protein_2.csv")
```
