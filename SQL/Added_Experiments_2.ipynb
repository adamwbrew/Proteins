```python
protiens_3 = protiens_3[['Protein Group', 'Protein Ids', 'Protein Names', 'Genes',
                        'First Protein Description', 'Peptides', 'Q331K_M1', 'Q331K_M2',
                        'WT_M3', 'WT_M4']]
protiens_3 = protiens_3.fillna(0)
protiens_3.info()
```

```python
# create mouse to protiens hit table based on protiens_2
protein_mouse_1 = protiens_3[["Protein_Names", "Q331K_M1"]]
protein_mouse_2 = protiens_3[["Protein_Names", "Q331K_M2"]]
protein_mouse_3 = protiens_3[["Protein_Names", "WT_M3"]]
protein_mouse_4 = protiens_3[["Protein_Names", "WT_M4"]]

protein_mouse_1["Sample_Name"] = "Q331K_M1"
protein_mouse_2["Sample_Name"] = "Q331K_M2"
protein_mouse_3["Sample_Name"] = "WT_M3"
protein_mouse_4["Sample_Name"] = "WT_M4"

protein_mouse_1 = protein_mouse_1.rename(columns = {"Q331K_M1":"Hits"})
protein_mouse_2 = protein_mouse_2.rename(columns = {"Q331K_M2":"Hits"})
protein_mouse_3 = protein_mouse_3.rename(columns = {"WT_M3":"Hits"})
protein_mouse_4 = protein_mouse_4.rename(columns = {"WT_M4":"Hits"})

union_dfs = pd.concat([protein_mouse_1, protein_mouse_2, protein_mouse_3, protein_mouse_4], ignore_index=True)

union_dfs = union_dfs[["Protein_Names", "Sample_Name", "Hits"]]
union_dfs.head()
```
